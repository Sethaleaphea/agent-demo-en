```python
import pandas as pd
kq = pd.read_csv('jiaoyuxue_data.csv', encoding='utf-8', index_col=0)
tmp = kq
tmp.head()
```

```python
#html 格式数据
import json
import html

def clear_txt(html_string):
    decoded_string = html.unescape(html_string)
    return decoded_string.replace('\r\n', '').replace('\n', '').replace('\r', '').replace(' ', '')

tmp['stem'] = tmp['stem'].apply(lambda x: clear_txt(x) if isinstance(x, str) else x)
tmp['analysis'] = tmp['analysis'].apply(lambda x: clear_txt(x) if isinstance(x, str) else x)
tmp['metas'] = tmp['metas'].apply(lambda x: clear_txt(x) if isinstance(x, str) else x)

import re
def remove_html_tags(s):
    if isinstance(s, str):
        return re.sub(r'<[^>]+>', '', s)  # 匹配所有 HTML 标签
    return s

def process_metas(metas):
    if metas is None:
        return None
    try:
        metas_list = json.loads(metas)
        metas_str = " ".join(f"{item['key']}: {item['value']}" for item in metas_list)
        # 去除 HTML 标签和多余字符
        metas_str = remove_html_tags(metas_str)
        metas_str = metas_str.replace('\\r\\n', '').replace('\r\n', '')
        return metas_str
    except json.JSONDecodeError:
        # 如果 metas 字段无法解析，返回原始值或标记错误
        return metas

# 更新 tmp 的 metas 列
tmp['metas'] = tmp['metas'].apply(process_metas)
tmp['analysis'] = tmp['analysis'].apply(lambda x: remove_html_tags(x) if isinstance(x, str) else x)
tmp['stem'] = tmp['stem'].apply(lambda x: remove_html_tags(x) if isinstance(x, str) else x)
tmp['answer'] = tmp['answer'].apply(lambda x: remove_html_tags(x) if isinstance(x, str) else x)

tmp.head()
```

```python
from trustrag.modules.retrieval.dense_retriever import DenseRetrieverConfig, DenseRetriever
from trustrag.modules.retrieval.embedding import FlagModelEmbedding

embedding_model_path = r"D:/Python/models/bge-large-zh-v1.5"  # 随便写一个就可以
index_path = r'D:\code\agent_demo\kg\index'

def RetrieverBuilder(embedding_model_path: str, index_path: str):
    retriever_config = DenseRetrieverConfig(
        model_name_or_path=embedding_model_path,
        dim=1024,
        index_path= index_path
    )
    embedding_generator = FlagModelEmbedding(embedding_model_path)
    retriever = DenseRetriever(retriever_config, embedding_generator)
    retriever.load_index(index_path)
    return retriever

retriever = RetrieverBuilder(embedding_model_path, index_path)
    
def RAG(query: str, retriever: DenseRetriever):
    contents = retriever.retrieve(query=query, top_k=3)
    # contents = '\n'.join([content['text'] for content in contents])
    return contents
```

```python
# 获取每个知识点的题目
# 用于存储每个 `oid` 对应的题目
oid_to_questions = {}
i = 0
tmp_counts = []
for item in point_result.iterrows():
    index, row = item
    oid = row['oid']
    pname = row['name']
    # 找出包含该 `oid` 的所有题目
    matching_rows = tmp[tmp['knowledge_point_ids'].astype(str).apply(lambda x: oid in x.split(","))]
    
    # 将结果保存到字典中
    oid_to_questions[oid] = matching_rows
    
    if len(matching_rows) > 0:
        # 查询实体 向量库
        print(f"oid: {oid} 对应的题目数量: {len(matching_rows)}")
        # 每个题目获取最相似的前五个实体
        related_list = []
        total_count = len(matching_rows)
        for _, question_row in matching_rows.iterrows():
            # print(f"题目: {question_row['stem']}")
            # print(f"分析: {question_row['analysis']}")
            # print(f"答案: {question_row['answer']}")
            # print(f"metas: {question_row['metas']}")
            if isinstance(question_row['metas'], str) == False:
                question_row['metas'] = ''
            if isinstance(question_row['analysis'], str) == False:
                question_row['analysis'] = ''
            try:
                # print(question_row['stem'] + question_row['metas'] + question_row['analysis'])
                contensts = RAG(question_row['stem'] + question_row['metas'] + question_row['answer'] + question_row['analysis'], retriever)
            except Exception as e:
                print(f"Error occurred: {e}")
                print(f"题目: {question_row['stem']}")
                print(f"分析: {question_row['analysis']}")
                print(f"答案: {question_row['answer']}")
                print(f"metas: {question_row['metas']}")
                total_count -= 1
                continue
            
            for content in contensts:
                # print(content)
                matches = re.findall(r'<<(.*?)>>', content['text'])
                related_list.append(matches[0])
                # print(f"oid: {oid} 对应的题目: {question_row['stem']}")
                # print(content['text'])
                
        
                # print(related_list)
        
        # 统计related_list 中的各实体出现次数
        related_counts = {}
        for related in related_list:
            if related in related_counts:
                related_counts[related] += 1
            else:
                related_counts[related] = 1
        # 按照出现次数排序
        sorted_related_counts = sorted(related_counts.items(), key=lambda x: x[1], reverse=True)
        # 打印排序后的结果
        for related, count in sorted_related_counts:
            print(f"oid: {oid} 对应的实体: {related} 出现次数: {count}")
        tmp_counts.append((oid, total_count, sorted_related_counts))

# 将结果保存到 json
import json
with open('tmp_counts.json', 'w', encoding='utf-8') as f:
    json.dump(tmp_counts, f, ensure_ascii=False, indent=4)
```

```python
strong = []
weak = []
normal = []
for oid, total_count, sorted_related_counts in tmp_counts:
    f1 = total_count * 0.04
    f2 = total_count * 0.1
    for related, count in sorted_related_counts:
        related = int(related)
        if count < f1:
            weak.append({
                'oid': oid,
                'related': related,
                'rate': count / total_count
            })
        elif count < f2:
            normal.append({
                'oid': oid,
                'related': related,
                'rate': count / total_count
            })
        else:
            strong.append({
                'oid': oid,
                'related': related,
                'rate': count / total_count
            })
```

```python
import pandas as pd
import ast

df_knowledge_points = pd.DataFrame(knowledge_points_data)

# 创建一个字典来存储每个知识点的题目列表
knowledge_points = {kp_id: [] for kp_id in df_knowledge_points['name']}

# 将题目根据知识点分发
for index, row in tmp.iterrows():
    point_ids = ast.literal_eval(row['knowledge_point_ids'])  # 将字符串转换为列表
    for point_id in point_ids:
        # print(point_id)
        try:
            knowledge_points[point_id].append(row)
        except KeyError:
            print(f"KeyError: {point_id} not found in knowledge points dictionary.")

# 输出分发后的结果
for point_id, questions in knowledge_points.items():
    print(f"知识点: {point_id}, 题目: {questions}")
```
