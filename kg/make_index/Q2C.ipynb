```python
point_result = pd.DataFrame(point_result)
```

```python
tmp['path'] = [[] for _ in range(len(tmp))]
```

```python
# 获取每个知识点的题目
# 用于存储每个 `oid` 对应的题目
oid_to_questions = {}
i = 0
tmp_counts = []
for item in point_result.iterrows():
    index, row = item
    oid = row['oid']
    pname = row['name']
    # 找出包含该 `oid` 的所有题目
    matching_rows = tmp[tmp['knowledge_point_ids'].astype(str).apply(lambda x: oid in x.split(","))]
    
    # 将结果保存到字典中
    oid_to_questions[oid] = matching_rows
    
    if len(matching_rows) > 0:
        # 查询实体 向量库
        # print(f"oid: {oid} 对应的题目数量: {len(matching_rows)}")
        # 每个题目获取最相似的前五个实体
        related_list = []
        total_count = len(matching_rows)
        for _, question_row in matching_rows.iterrows():
            # print(f"题目: {question_row['stem']}")
            # print(f"分析: {question_row['analysis']}")
            # print(f"答案: {question_row['answer']}")
            # print(f"metas: {question_row['metas']}")
            if isinstance(question_row['metas'], str) == False:
                question_row['metas'] = ''
            if isinstance(question_row['analysis'], str) == False:
                question_row['analysis'] = ''
            try:
                # print(question_row['stem'] + question_row['metas'] + question_row['analysis'])
                query = question_row['stem'] + question_row['metas'] + question_row['answer'] + question_row['analysis']
                contensts = vector_store_loaded.similarity_search(query, k=3)
            except Exception as e:
                print(f"Error occurred: {e}")
                print(f"题目: {question_row['stem']}")
                print(f"分析: {question_row['analysis']}")
                print(f"答案: {question_row['answer']}")
                print(f"metas: {question_row['metas']}")
                total_count -= 1
                continue
            
            print(query)
            for content in contensts:
                print(content.page_content[:100])
                # print(content.metadata)
                related_list.append(content.metadata['source'])
                tmp_index = question_row.name
                tmp.at[tmp_index, 'path'].append(content.metadata['path'])
            # question_row['path'] = tmp_related

        
        # 统计related_list 中的各实体出现次数
        related_counts = {}
        for related in related_list:
            if related in related_counts:
                related_counts[related] += 1
            else:
                related_counts[related] = 1
        # 按照出现次数排序
        sorted_related_counts = sorted(related_counts.items(), key=lambda x: x[1], reverse=True)
        # 打印排序后的结果
        for related, count in sorted_related_counts:
            print(f"oid: {oid} 对应的实体: {related} 出现次数: {count}")
        tmp_counts.append((oid, total_count, sorted_related_counts))

# 将结果保存到 json
import json
with open('tmp_counts.json', 'w', encoding='utf-8') as f:
    json.dump(tmp_counts, f, ensure_ascii=False, indent=4)
```
